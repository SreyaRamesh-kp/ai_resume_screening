```python
!pip install sentence-transformers
!pip install pymupdf
!pip install scikit-learn
import joblib

import fitz


from google.colab import files
uploaded=files.upload()
filename = list(uploaded.keys())[0]
doc=fitz.open(filename)

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

text = ""

for page in doc:
    text += page.get_text()

# print(text)

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

save_model=joblib.dump(model,'model.pkl')
job_description = """
Looking for an AI & Data Science Engineer.

Required Skills:
Python
Machine Learning
Deep Learning
SQL
Power BI
Flask
"""


resume_embedding = model.encode(text)
job_embedding = model.encode(job_description)


score = cosine_similarity(
    [resume_embedding],
    [job_embedding]
)


match_percentage = score[0][0] * 100


```

```python
skills = [
    "Python",
    "Machine Learning",
    "Deep Learning",
    "SQL",
    "Power BI",
    "Flask",
    "TensorFlow",
    "PyTorch",
    "Pandas",
    "NumPy",
    "Scikit-learn",
    "Data Analysis",
    "Data Visualization",
    "Git",
    "Docker",
    "AWS"
]
resume_skill=[]
for skill in skills:
  if skill.lower() in text.lower():

    resume_skill.append(skill.lower())
# print("reume skill")
# print("----"*10)
# print(resume_skill)


job_skill=[]

for skill in skills:
  if skill.lower() in job_description.lower():
    job_skill.append(skill.lower())

# print("Job skill")
# print("---"*10)
# print(job_skill)


match_skill=[]

for skill in resume_skill:
  if skill in job_skill:
    match_skill.append(skill)




missing_skill=[]

for skill in job_skill:
  if skill not in resume_skill:
    missing_skill.append(skill)



print(f"Resume Match: {match_percentage:.2f}%  , uccessfully completed the basic AI Resume Screening System.")
print("Match skill")
print("---"*10)
print("✅",match_skill)

print("Missing_skill")
print("---"*19)
print("❌",missing_skill)
```
